**Задание 1.**

Написать функцию отображения элементов одного множества в другое. Первое множество - художники, второе множество - картины художников.
Множество художников - Серов, Левитан, Репин, Васнецов, Шишкин. 
Множество картин - "Утро в сосновом бору", "Богатыри", "Золотая осень", "Девочка с персиками", "Запорожцы", "Буралки на Волге", "Март", "Алёнушка". 


```python
PAINTERS = {
    'SEROV': 'Серов',
    'LEVITAN': 'Левитан',
    'REPIN': 'Репин',
    'VASNECOV': 'Васнецов',
    'SHISHKIN': 'Шишкин',
}

PICTURES = {
    PAINTERS['SEROV']: ['Девочка с персиками'],
    PAINTERS['LEVITAN']: ['Золотая осень', 'Март'],
    PAINTERS['REPIN']: ['Запорожцы', 'Буралки на Волге'],
    PAINTERS['VASNECOV']: ['Богатыри', 'Алёнушка'],
    PAINTERS['SHISHKIN']: ['Утро в сосновом бору'],
}

def pictures_by_painters(painters, pictures):
    for painter in painters.values():
        painter_pictures = pictures.get(painter)
        
        if painter_pictures:
            print(f'Автор: {painter}')
            print(f'Картины: {", ".join(painter_pictures)}')
        print()
            
pictures_by_painters(PAINTERS, PICTURES)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def show_plot(f):
    x = np.linspace(-5, 5, 100)
    y = [f(x[i]) for i in range(len(x))]

    plt.plot(x, y)
    plt.title('График функции')
    plt.ylabel('y')
    plt.xlabel('x')

    plt.show()
```

```python
def f1(x):
    return (3*x/4) - 1

show_plot(f1)
```

```python
def f2(x):
    return (3*x/4) + 2

show_plot(f2)
```

```python
def f3(x):
    return 2 - 2*x

show_plot(f3)
```

```python
def f4(x):
    return 2 + 2*x

show_plot(f4)
```

```python
def f5(x):
    return ((x + 1)**2)/4 - 2

show_plot(f5)
```

```python
def f6(x):
    return ((x - 1)**2)/4 - 2

show_plot(f6)
```

```python
def f7(x):
    return (x - 1)*(x - 3)

show_plot(f7)
```

```python
def f8(x):
    return (x + 1)*(x + 3)

show_plot(f8)
```

```python
def f9(x):
    return 2 - x**2

show_plot(f9)
```

```python
def f10(x):
    return 2 - x**2/4

show_plot(f10)
```

### 6. Выдвинуть несколько гипотез по набору имеющихся переменных и проверить эти гипотезы с использованием t-test.

### Гипотеза-1: "Атмосферные двигатели по трассе расходуют столько же сколько турбированные"

```python
atmo_hw=df[df["aspiration"]=="std"]["highway-mpg"]
turbo_hw=df[df["aspiration"]=="turbo"]["highway-mpg"]
```

```python
len(atmo_hw), len(turbo_hw)
```

```python
atmo_hw = atmo_hw.sample(35)
turbo_hw = turbo_hw.sample(35)
```

```python
statistics.variance(atmo_hw), statistics.variance(turbo_hw)
```

```python
sts.ttest_ind(atmo_hw, turbo_hw, equal_var=False)
```

### Гипотеза-2: "Атмосферные двигатели по городу расходуют столько же сколько турбированные"

```python
atmo_city=df[df["aspiration"]=="std"]["city-mpg"]
turbo_city=df[df["aspiration"]=="turbo"]["city-mpg"]
```

```python
len(atmo_city), len(turbo_city)
```

```python
atmo_city = atmo_city.sample(35)
turbo_city = turbo_city.sample(35)
```

```python
statistics.variance(atmo_city), statistics.variance(turbo_city)
```

```python
sts.ttest_ind(atmo_city, turbo_city, equal_var=False)
```

### По образу практики, попробуйте создать искусственный датасет с лишними столбцами.

Для начала создам датафрейм, содержащий инофрмацию о грешниках:

Возраст - 18 - 90 age

Пол - m(male) / f(female)

Вес - 40 - 210 kg

Рост - 150 - 225 sm

По данному датафрейму, буду пытаться понять влияет количество плохих помыслов, с возрастом)).

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
```

```python
n_samples = 500
YES = 'Y'
NO = 'N'
MALE = 'M'
FEMALE = 'F'

age = np.random.choice(88, n_samples) + 3
sex = [FEMALE if i > 1 else MALE for i in np.random.choice(3, n_samples)]
weight = np.random.choice(80, n_samples) + 10
height = np.random.choice(160, n_samples) + 50


amount_of_bad_thoughts = np.random.choice(30, n_samples) + n_samples * 2
```

```python
data = pd.DataFrame({'age': age, 'sex': sex, 'weight': weight,
                     'height': height, 'amount_of_bad_thoughts': amount_of_bad_thoughts})
data.head(10)
```

```python
from sklearn.metrics import mean_absolute_error

x = data[['amount_of_bad_thoughts', 'height', 'weight']] ## плохие помыслы, рост и вес 
y = data['age'] ## с взрастом
reg = LinearRegression().fit(x, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['amount_of_bad_thoughts', 'height', 'weight']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
from sklearn.metrics import mean_absolute_error

x = data[['amount_of_bad_thoughts', 'weight']] ## плохие помыслы и вес
y = data['age'] ## с взрастом
reg = LinearRegression().fit(x, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['amount_of_bad_thoughts', 'weight']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
# Создаем новый признак
# amount of good deeds - количество хороших дел
data['amount_of_good_deeds'] = data['age'] * data['height']
data.head(10)
```

```python
x = data[['amount_of_good_deeds']] ## хорошие поступки
y = data['age'] ## с возратом
reg = LinearRegression().fit(x, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['age']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```
